```python
#import pandas and other libraries
import pandas as pd
import numpy as np
import json
from pprint import pprint
```

```python
#load json file
with open('nflData.json') as f:
    data = json.load(f)
```

```python
#convert to pandas dataframe
nflFrame = pd.DataFrame(data)
nflFrame.head()
```

```python
#Drop season and week columns
playerInfo = nflFrame.drop("season", axis=1)
for i in range(1,14):
    week = f"week_{i}"
    playerInfo.drop(week, axis=1, inplace=True)
playerInfo
playerInfo["Fantasy Points"] = playerInfo["Fantasy Points"].round(2)
playerInfo
```

```python
#Export Player Info to CSV
playerInfo.to_csv("csvs/player_Info", index=False, header=True)
```

```python
seasonStats = nflFrame["season"]
offenseDf = nflFrame[nflFrame["position_type"] == "O"].reset_index(drop=True)
defenseDf = nflFrame[nflFrame["position_type"] == "DT"].reset_index(drop=True)
kickerDf = nflFrame[nflFrame["position_type"] == "K"].reset_index(drop=True)
```

```python
#Extrtact table for Offense Stats
offSeasonStats = offenseDf["season"]
# Step 1: Explode the 'season' column to separate list elements
nflFrame_exploded = offenseDf.explode("season")

# Step 2: Flatten the dictionaries in the 'season' column
offSeasonStats = pd.json_normalize(nflFrame_exploded["season"])

# Step 3: Add the player_id back to the flattened DataFrame
offSeasonStats["player_id"] = nflFrame_exploded["player_id"]
offSeasonStats["Fantasy Points"] = offSeasonStats["Fantasy Points"].round(2)
offSeasonStats
```

```python
#Extrtact table for Offense Stats
defSeasonStats = defenseDf["season"]
# Step 1: Explode the 'season' column to separate list elements
defFrame_exploded = defenseDf.explode("season")

# Step 2: Flatten the dictionaries in the 'season' column
defSeasonStats = pd.json_normalize(defFrame_exploded["season"])

# Step 3: Add the player_id back to the flattened DataFrame
defSeasonStats["player_id"] = defFrame_exploded["player_id"]
defSeasonStats["Fantasy Points"] = defSeasonStats["Fantasy Points"].round(2)
defSeasonStats
```

```python
#Extrtact table for Offense Stats
kickSeasonStats = kickerDf["season"]
# Step 1: Explode the 'season' column to separate list elements
kickFrame_exploded = kickerDf.explode("season")

# Step 2: Flatten the dictionaries in the 'season' column
kickSeasonStats = pd.json_normalize(kickFrame_exploded["season"])

# Step 3: Add the player_id back to the flattened DataFrame
kickSeasonStats["player_id"] = kickFrame_exploded["player_id"]
kickSeasonStats["Fantasy Points"] = kickSeasonStats["Fantasy Points"].round(2)
kickSeasonStats
```

```python
#Export offense, defense, and kicker stats to CSV
offSeasonStats.to_csv("csvs/offense_season_stats", index=False, header=True)
defSeasonStats.to_csv("csvs/defense_season_stats", index=False, header=True)
kickSeasonStats.to_csv("csvs/kicker_season_stats", index=False, header=True)
```
